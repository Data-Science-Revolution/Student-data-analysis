```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df_reg = pd.read_csv(r'E:\1_Praxis Data Science\TERM 1\STS Term 1\Project\Regression_1.csv',sep=",")
```

```python
df_reg
```

```python
print(df_reg.corr())
```

```python
X1 = df_reg['POP_below_Poverty_Line']
y1 = df_reg['Total_Students_2020']
```

```python
import seaborn as ana
fig, axs = plt.subplots(2, figsize = (5,5))
plt1 = sns.boxplot(df_reg['Total_Students_2020'], ax = axs[0])
plt2 = sns.boxplot(df_reg['POP_below_Poverty_Line'], ax = axs[1])
plt.tight_layout()
```

```python
sns.pairplot(df_reg, x_vars=['POP_below_Poverty_Line'], y_vars=['Total_Students_2020'], height=4, aspect=1, kind='scatter')
plt.show()
```

```python
msk = np.random.rand(len(df_reg)) < 0.8
train = df_reg[msk]
test = df_reg[~msk]
```

```python
print(test)
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['POP_below_Poverty_Line']])
train_y = np.asanyarray(train[['Total_Students_2020']])
regr.fit(train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train.POP_below_Poverty_Line,train.Total_Students_2020,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("POP_below_Poverty_Line")
plt.ylabel("Total_Students_2020")
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['POP_below_Poverty_Line']])
test_y = np.asanyarray(test[['Total_Students_2020']])
test_y_ = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y , test_y_) )
```
